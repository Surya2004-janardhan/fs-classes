Linked list - Linear data structure where each element (node) points to the next (node),
allowing for efficient insertion and deletion. It consists of nodes that contain 
data and a reference to the next node in the sequence.

```python
# linked list class 

# | prev | data | next |

class Node:
    def __init__(self, value):
        self.prev = None
        self.data = value
        self.next = None

# obj for node


#  | 10 | None |
node1 = Node(10) # creating node with value 10 and next = none 
node2 = Node(10) 
node3 = Node(30)
node4 = Node(40)

# linked list u cant accesss in btw nodes - first node is called as head node 

# doubly linked list - each node has 3 parts - prev, data, next
# singly linked list - each node has 2 parts - data, next

# putting together doubly linked list
node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

node4.next = node1
node1.prev = node4






```

```python
print(node1.next.next.data)
```

```python
# traversing linked list - accessing each node one by one
head = node1
current = head

seen = set()

while current is not None:
    
    if current in seen:
        break
    seen.add(current)
    
    prev_value = current.prev.data if current.prev else None
    next_value = current.next.data if current.next else None
    print(prev_value ,current.data, next_value ) 




    
    current = current.next
print(seen)


# # now current store last node lets do reverse traversal
# tail  = node4

# current = tail
# while current is not None:
#     prev_value = current.prev.data if current.prev else None
#     next_value = current.next.data if current.next else None
#     print(prev_value ,current.data, next_value ) 
#     current = current.prev
```
